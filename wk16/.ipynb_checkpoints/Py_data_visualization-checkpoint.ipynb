```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
plt.plot(np.random.randn(10) * 10) 

ax2 = fig.add_subplot(2,2,2)
plt.plot(np.random.random(10) * 10)

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randint(1, 100, size=10),'k. ') 

ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.randn(10) * 50, 'k--')  
```

```python
fig = plt.figure()
dataset = randn(50).cumsum()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
ax1.plot(dataset, 'k--')
ax2.hist(dataset, bins=20, color='k', alpha=0.25)
ax3.scatter(dataset,dataset.cumsum())
ax4.plot(dataset,'ko--')
fig
```

```python
data = randn(30).cumsum()
plt.plot(randn(30).cumsum(), 'k-',drawstyle = 'steps-post')
plt.plot(randn(30).cumsum(), 'k--')
```

```python
# using lables 
plt.plot(data, 'k-', drawstyle='steps-post', label='step-post')
plt.plot(data, 'k--', label='dashed-line')
plt.legend(loc='best') #place a legned on the axes.locs = 
```

```python
#plot multiple columns 
data = {
    'col1': np.random.randint(1,21, size=20),
    'col2': np.random.randint(20,41, size=20),
    'col3': randn(20),
    'col4': randn(20).cumsum()
    
}
df = DataFrame(data)
df
```

```python
plt.plot(df)
```

```python
plt.plot(df['col1'],'k.', label='col1')
plt.plot(df['col2'],'ko', label='col2')
plt.legend(loc='best')
```

```python
#ticks, labels, and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
#ticks
ax.set_xticks(list(range(0,1001,250)))


#labels
ax.set_xticklabels(['one','two','three','four','five'], rotation = 30, fontsize='small' )

#titles
ax.set_title('My first matplotlib plot with ticks')

#labels for x y 
ax.set_xlabel('stages')
ax.set_ylabel('values')


fig
```

```python
#use dictinary for title, xlable, and ylable 
props = {'title': 'My First matplotlip plot with ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}
ax.set(**props)
fig
```
